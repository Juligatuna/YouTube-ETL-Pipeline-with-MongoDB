```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient("mongodb://localhost:27017")
db = client['youtube_etl']
collection = db['videos']

df = pd.DataFrame(list(collection.find()))
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Top channels by number of videos

```python
top_channels = df['channelTitle'].sort_values().head(5)
sns.barplot(x=top_channels.values, y=top_channels.index)
plt.title("Top 5 Channels by video count")
plt.ylabel("No. of videos")
plt.xticks(rotation=45)
plt.show()
```

# 2. Distribution of publish dates

```python
df['published_at'] = pd.to_datetime(df['published_at'])
df['published_at'].hist(bins = 20, figsize=(10,5))
plt.title('Video Publish Dates Distribution')
plt.show()
```
